```python
import pandas as pd
from sklearn.model_selection import train_test_split
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
data = pd.read_csv('Testing.csv')
data.head()
```

```python
#performing EDA on 'test'.
data.info()

data.describe()
```

```python
label = 'prognosis'
```

```python
data[label].describe()
```

```python
# Split the data into training and testing sets.
data1 = TabularDataset("Testing.csv")

train_data, test_data = train_test_split(data1, test_size=0.4)

# Create a TabularPredictor object for making precitions or inferences.
predictor = TabularPredictor(label="prognosis", eval_metric = "root_mean_squared_error").fit(
    train_data=train_data, time_limit=600, presets="best_quality"
)
#Tuned the hyperparameters of 'time_limit' which specifies the time for whihc the model is supposed to be trained for.
#Similarly, the 'presets' hyperparameter determines the kinda algorithm we want AutoML to implement to make future predictions based on the 'quality'
#of output from the models it uses while training on the training dataset...
#supported hyperparameter documentation: https://docs.aws.amazon.com/sagemaker/latest/dg/autogluon-tabular-hyperparameters.html
```

```python
#printing the summary of trianing and models used.
predictor.fit_summary()
```

```python
# Make predictions on the test data
predictions = predictor.predict(test_data)

# Evaluate the predictions
#print(predictor.evaluate_predictions(y_true=test_data["prognosis"], y_pred=predictions))

predictions.describe()
```
